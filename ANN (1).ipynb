Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

Loading the dataset

```python
data = pd.read_csv('/Alphabets_data.csv')
data.head()
```

```python
#remove duplicate data
data = data.drop_duplicates()
```

```python
data.duplicated().sum()
```

Scaling the data

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
X= data.drop(columns=['letter'])
y = data['letter']
```

```python
#mimax scaler
scaler = MinMaxScaler()
scaled_data= scaler.fit_transform(X)
pd.DataFrame(scaled_data)
```

Applying OneHotEncoder

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
# One-hot encode the target variable
OHE= OneHotEncoder()
encoded_data= OHE.fit_transform(y.values.reshape(-1, 1))
encoded_df= pd.DataFrame(encoded_data.toarray())
```

```python
encoded_df
```

Spliting the dataset

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(scaled_data,encoded_df,test_size=0.2,random_state=42)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model1 = Sequential([
    Dense(12, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(26, activation='softmax')
])
```

```python
model1.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
training = model1.fit(X_train, encoded_df, epochs=20, batch_size=20, validation_split=0.2)
```

```python
model2 = Sequential([
    Dense(56, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(26, activation='sigmoid')
])
```

```python
model2.compile(optimizer='RMSprop', loss='categorical_crossentropy', metrics=['accuracy'])
training = model2.fit(X_train,encoded_df, epochs=20, batch_size=42, validation_split=0.2)
```

```python
model3 = Sequential([
    Dense(56, activation='linear', input_shape=(X_train.shape[1],)),
    Dense(26, activation='sigmoid')
])
```

```python
model3.compile(optimizer='RMSprop', loss='categorical_crossentropy', metrics=['accuracy'])
training = model3.fit(X_train, encoded_df, epochs=20, batch_size=42, validation_split=0.2)
```

```python
model4 = Sequential([
    Dense(56, activation='tanh', input_shape=(X_train.shape[1],)),
    Dense(26, activation='sigmoid')
])
```

```python
model4.compile(optimizer='ADAgrad', loss='categorical_crossentropy', metrics=['accuracy'])
training = model4.fit(X_train, encoded_df, epochs=20, batch_size=42, validation_split=0.2)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import numpy as np
```

```python
# Get the true labels
y_true = np.argmax(y_test, axis=1)

# Get the predicted labels
y_pred = np.argmax(model1.predict(X_test), axis=1)

accuracy = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred, average='weighted')
recall = recall_score(y_true, y_pred, average='weighted')
f1 = f1_score(y_true, y_pred, average='weighted')
conf_matrix = confusion_matrix(y_true, y_pred)

print("Model with relu , softmax and adam optimizer ")
print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1-score: {f1}")
```

```python
y_true = np.argmax(y_test, axis=1)

# Get the predicted labels
y_pred = np.argmax(model2.predict(X_test), axis=1)

accuracy = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred, average='weighted')
recall = recall_score(y_true, y_pred, average='weighted')
f1 = f1_score(y_true, y_pred, average='weighted')
conf_matrix = confusion_matrix(y_true, y_pred)

print("Model with relu , sigmoid and RMSprop")
print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1-score: {f1}")
```
